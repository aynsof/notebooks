```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1,2])
y_train = np.array([300, 500])
```

```python
print(x_train)
```

```python
m = x_train.shape[0]
print(m)
```

```python
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title("Housing Prices")
plt.ylabel("Price (in 1000s of dollars)")
plt.xlabel("Size (1000 sqft)")
plt.show()
```

```python
w = 100
b = 100
```

```python
def compute_model_output(w, x, b):
    m = x.shape[0]
    
    f_wb = np.zeros(m)
    
    for i in range(m):
        f_wb[i] = (w * x[i]) + b
        print(f_wb[i])
    
    return f_wb
```

```python
w = 200
b = 100

f_wb = compute_model_output(w, x_train, b)

plt.plot(x_train, f_wb, c='b', label='Prediction')
plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Data')

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')

plt.show()
```
